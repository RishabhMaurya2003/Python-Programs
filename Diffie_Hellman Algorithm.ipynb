```python
p = int(input("Enter a prime number p: "))
g = int(input("Enter g: "))
while (1):
    xa = int(input("Enter xa:")) 
    if xa<p :
        break
while(1):
  xb = int(input("Enter xb: "))
  if xb<p:
    break

YA = g ** xa % p        
YB = g ** xb % p

KA = YB ** xa % p
KB = YA ** xb % p

if KA==KB:
  print("Successful")
```
