```python
Rishabh Maurya
def hebbian_learning (samples):
    print(f'{"INPUT":^8} {"TARGET":^16} {"WEIGHT CHANGES":^15}{"WEIGHTS":^25}')
    w1, w2, b = 0, 0, 0
    print(' ' * 45, f'({w1:2}, {w2:2}, {b:2})')
    for x1, x2, y in samples:
        w1= w1 + x1*y
        w2= w2 + x2*y
        b = b+ y
        print(f'({x1:2}, {x2:2}) {y:2} ({x1*y:2}, {x2*y:2}, {y:2}) ({w1:2}, {w2:2}, {b:2})')
```
